# **Modelling**

```python
from sklearn import preprocessing
def label_encoder(data):
    le = preprocessing.LabelEncoder()
    data = le.fit_transform(data)
    return data
```

```python
datatrain['No Transaksi Tabular'] = label_encoder(datatrain['NO SPB'])
datatrain.head(15)
```

```python
Basket = (datatrain[datatrain['tahun']==2021]
          .groupby(['NO SPB', 'NAMA BARANG'])['QTY']
          .sum().unstack().reset_index().fillna(0)
          .set_index('NO SPB'))

Basket.head()
```

```python
def sum_to_boolean(x):
    if x<=0:
        return 0
    else:
        return 1

Basket_Final = Basket.applymap(sum_to_boolean)
```

```python
Basket_Final.head()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
Frequent_itemsets = apriori(Basket_Final, min_support = 0.04, use_colnames = True)
Frequent_itemsets['length'] = Frequent_itemsets['itemsets'].apply(lambda x: len(x))
Frequent_itemsets.tail()
```

```python
Asso_Rules = association_rules(Frequent_itemsets, metric = "lift", min_threshold = 1)
Asso_Rules.sort_values('lift',ascending = False).head()
```

```python
Asso_Rules = association_rules(Frequent_itemsets, metric = "lift", min_threshold = 1)
Asso_Rules.sort_values(['lift'],ascending = [False])
```

```python
Asso_RulesFiltered = Asso_Rules[(Asso_Rules['lift'] >= 11)&(Asso_Rules ['confidence'] >= 0.6)].sort_values(['lift'],ascending = [False])
Asso_RulesFiltered
```

```python
Asso_RulesFiltered = Asso_Rules[(Asso_Rules ['confidence'] >= 0.6)].sort_values(['lift'],ascending = [False])
Asso_RulesFiltered.sort_values(['confidence'],ascending = [False])
```
